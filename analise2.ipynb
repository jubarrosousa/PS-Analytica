Criei uma coluna a mais no dataframe com a média de "NaN" encontrada para cada escola, e plotei um gráfico para cada tipo de escola.

Observando os gráficos, é possível perceber que o dataset possui apenas dados consideráveis para escolas estaduais e para algumas federais, e não parece ter relação entre a falta de dados e a localização das escolas nas regiões rurais e urbanas. 

Tirei todas as escolas que não possuiam nenhum dado além dos ids, tipo e localização, e analisei quais informações as escolas mais tinham.

```python
df= regiao.copy()

# retirei as colunas que todas as escolas apresentavam dados
df = df.drop(['id_escola', 'id_municipio', 'localizacao', 'rede'], axis = 1)

df = df.dropna(how='all')

df = df.T
df['% NaN'] = (df.isnull().sum(axis=1))/(len(df.columns))

y = df['% NaN']
print(y)
x = np.arange(len(df.index))

# Os 5 índices com mais e menos dados
print(df.nsmallest(5, '% NaN')['% NaN'])
print()
print(df.nlargest(10, '% NaN')['% NaN'])


figure(figsize=(8, 6), dpi=80)
plt.bar(x, y, align = 'center')
plt.xlabel('Índices')
plt.ylabel('% NaN')

plt.show()
```

Ao fim dessa análise, é visto menos de 30% das escolas continham pelo menos algum dado para os índices escolhidos.
